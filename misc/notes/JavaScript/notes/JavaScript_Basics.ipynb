JavaScript Basics
==========



## Four Basic Data Types

### Nums
* Catch-all for (most) Integers, Floats, Doubles
* `4`
* `-10.5`

### Strings 
* Double **OR** single quotes.
* `'Hello'`
* `"Hello"`
* `"hello".length`

### Booleans
* `true` and `false`

### Null vs Undefined
* `null`
  * literally nothing
  
* `undefined`
  * something has no value, yet

### All variables in JS are Dynamically Typed
* `var name = "Rusty";`
* `var name = 73;`
* `var name = true; `


### Starter functions
* `alert()`
  *  Gives annoying pop-up.
  
* `console.log()`
  * Prints to JS console only.

* `prompt()`
  * Prompts the user for input  (kinda like scanf())


### Comparisons
* `==` and `!=`
  * Compares only value (type coercion)

* `===` and `!==`
 * Compares value AND type (preferred)

## Higher Order Functions

### Higher Order Function
  * a function that either takea other functions as arguments, or returns a function.


Example:
* `setInterval()` takes in 2 args:
  * a function pointer (function name)
  * interval (milliseconds)
  
` function sing() {
  console.log("Twinkle Twinkle");
  console.log("How I wonder");
 }`

`setInterval(sing, 1000);`

<hr>


#### Alt format: defining the function inline.
  * Anonymous func, notice the func doesn't have a name (will never be used outside of setInterval()).

`setInterval(function (){
  console.log("Heyyyyyy, yahhh");
  console.log("Heyyyyyy, yahhh!");
}, 2000);`



**NOTE:**
* setInterval() returns a number, "num"
  * the number is needed for stopping setInterval()


* To stop:
  * clearInterval(num)


## Arrays

### Add to end of array, return length of array
`array.push()`

### Remove from end of array, return removed item.
`array.pop()`

### Add to beginning of array
`array.unshift()`

### Remove from beginning of array.
`array.shift()`

### Where a string pattern starts
`array.indexOf()`

### Select a slice of an array `[start, end)`
`array.slice(start, end)`

### Foreach (higher-order)
`array.forEach()`

* Example: Iterate through and print every "thing" in an array.

   ` var colors = ["red", "orange", "yellow", "green"];`

   ` colors.forEach(function(color) { 
     console.log(color) 
  });`
  
### Remove from specified index
*  `array.splice(index, numItemsToDeleteAfterIndex)`

## Objects
* (Basically Dictionaries)

### Declaration

Example (most common):<br> 
   `var person = {
    name: "Kathy",
    age: 21,
    city: "Orlando"
};`



Example 2 (reltively common):<br>
   `var person = {};`<br>
   `name: "Kathy";
    age: 21;
    city: "Orlando";
    `
    
Example 3 (least common):<br>
   `var person = new Object();`<br>
   `var person = {
    name: "Kathy",
    age: 21,
    city: "Orlando"
};`

#### Data in Objects can be accessed with either dot notation or bracket notation
  * `console.log(person.name);` --> `"Kathy"`
  * `console.log(person["name"]);` --> `"Kathy"`
  
  
  * When to Use Bracket Notation
     - When needing to access an object's field using a variable name.
     - When field/property in an object starts with a number.
     - When needing to access an object's field/prop and the field has spaces.
     

  * When to use Dot Notation
    - Every other time.
