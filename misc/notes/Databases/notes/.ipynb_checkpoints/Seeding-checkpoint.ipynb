# Seeding

## What Is Seeding?
## Seeding refers to planting sample data in a database.
* Uses:
  * More of a sanity check when debugging 
  * Making sure changes propogate quickly (and are able to be seen/noticed immediately)

## seeds file:
  * Runs every time the server is started
  * Clears db
  * Repropogates DB with planted data   
    
### Construction of seeds file:
  * create a file named "seeds.js"
  * `require("./seeds.js")` from your main app.js file.

### Configure seeds.js like so:
```js
    var mongoose   = require("mongoose");
    var Campground = require("./models/campground");
    var Comment    = require("./models/comment);

    var data = [
        {
            name: "Seed/Sample Name Goes Here",
            url: "Seed/Sample URL Goes Here",
            description: "Seed/Sample Description Goes Here"
        }
    ];

    function seedDB(){
   //Remove all campgrounds
   Campground.remove({}, function(err){
        if(err){
            console.log(err);
        }
        console.log("removed campgrounds!");
         //add a few campgrounds
        data.forEach(function(seed){
            Campground.create(seed, function(err, campground){
                if(err){
                    console.log(err)
                } else {
                    console.log("added a campground");
                    //create a comment
                    Comment.create(
                        {
                            text: "This place is great, but I wish there was internet",
                            author: "Homer"
                        }, function(err, comment){
                            if(err){
                                console.log(err);
                            } else {
                                campground.comments.push(comment);
                                campground.save();
                                console.log("Created new comment");
                            }
                        });
                }
            });
        });
    }); 
    //add a few comments
}

module.exports = seedDB;
```
