## Cats.js



```javascript
  
    // include mongoose and start db.
    var mongoose = require("mongoose");
    mongoose.connect("mongodb://localhost/cat_app");

    // Design the "class"
    var catSchema = mongoose.Schema({
      name: String,
      age: Number,
      temperament: String
    });

    // Set up the "constructor" with info from the schema.
    var Cat = mongoose.model("Cat", catSchema);

    // call "constructor", make a new Cat.
    var george = new Cat({
      name: "George",
      age: 11,
      temperament: "Grumpy"
    });


    // insert the cat into the DB, pass a callback to handle errors.
    // callback takes:
    //   the error code, and
    //   whatever is returned from the db (a cat in this example.)
    george.save(function(err, cat){
      if (err) {
        console.log("SOOMETHING'S WRONG!");

      }else{
        console.log("WE SAVED A CAT to DB:");
        console.log(cat);
      }
    })

    // make a new cat
    //   * pass the JSON representing the cat as the first arg inline.
    // and have/insert it to the db all in one,
    //
    // pass a callback to handle errors.
    //
    // callback takes:
    //   the error code, and
    //   whatever is returned from the db (a cat in this example.)
    Cat.create({

        name: "Snow White",
        age: 15,
        temperament: "Nice"
    }, function(err, cat) {
        if (err) {

          console.log("SOOMETHING'S WRONG!");
        }else{

          console.log("WE SAVED A CAT to DB");
          console.log(cat);
        }
    });

    // look for a cat matching the JSON specified in arg1, (ALL in this ex)
    // pass a callback to handle errors.
    // callback takes:
    //   the error code, and
    //   whatever is returned from the db (a cat in this example.)
    // print the cats
    Cat.find({}, function(err, cats) {
      if (err) {
        console.log("ERROR IN .find()!");

      }else{
        console.log("ALL THE CATS!");
        console.log(cats);
      }
    });
```
