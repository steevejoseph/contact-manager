## Internet Basics
1. Site typed in adress bar
2. Query submitted to ISP
3. Within ISP, DNS takes Domain name and turns it into unique IP address (unique locator for server)
4. New HTTP request (with hella info like client IP, timestamp, specific request (webpage), type of request)
5. Site's server figures out what is being asked for
6. Server builds the right content, sometimes pulling info from DBs
7. Server responds with HTML/CSS/JS
8. Response sent (raw)
9. Browser renders


## Static v Dynamic Sites
* Dynamic:
  * Sites that are compiled on the server side.
  * Server pulls together/constructs a webpage before sending content back as a response.
  * Possibly different HTML/CSS/JS sent back as a response, every time a request comes up.
  * Prolly gonna have a DB.

* Static
  * NOT dynamic
  * Exactly the SAME content every time.

### Flow
*Frontend*, **Backend**

*1. Ask for reddit homepage*

**2. Get top posts from DB, sent back home page content**

*3. Browser renders homepage*

*4. User enters "dogs" in search box and submits form*

**5. Finds all posts in DB about "dogs"**

**6. Sends back HTML for search results page**

*7. Browser renders search page*
