## Intro to Express

### What is a framework? How is it different from a library?
**[Good SO answer](https://stackoverflow.com/questions/3057526/framework-vs-toolkit-vs-library)**
  > The most important (and the defining) difference between a library and a franework is Inversion of Control. Library: you call the library, Framework:  the framework calls you.
  
**Library:**
  * Collection of functionality that you can call/are in control of.
  
**Framework** 
  * All control-flow is already defined. 
  * Has pre-defined witespots where your own code will go.

## Routing Basics
```javascript
// load up express
var express = require("express");

// bring forth all of express's methods/var/etc
// ALWAYS saved to a var, USUALLY called "app"
var app = express()


// First route:
// "/" => "Hi!"
// "/bye" => "Bye!"
// "/dog" => "MEOW!"

// Syntax for get requests:
// app.get("url", function(request, response){});
// param1: url/path
// param2: callback function
//    takes in2 params:
//      req[uest], res[ponse] 
//      req and res are objects that hold all info about the request and response, respectively.
app.get("/", function(req, res){
  
  // Send a respoonse saying hi.
  res.send("Hi there!");
});

app.get("/bye", function(req, res) {
  res.send("Bye!");
})

app.get("/dog", function(req, res){
  res.send("MEOW!");
});

// Respond by welcoming the user to the specific subredditName they submitted.
// ".../r/puppies" --> "WELCOME TO THE PUPPIES SUBREDDIT."
app.get("/r/:subredditName", function(req, res){
  
  // req and req.params is in JSON.
  var subreddit = req.params.subredditName;
  
  res.send("WELCOME TO THE " + subreddit.toUpperCase() + " SUBREDDIT!" );
});

// Splat/star that is returned any time a GET request is processed for a route
//  that has not been defined or is not expected.
//  (USUALLY ERROR 404)
// THIS LINE MUST BE LAST, since star matches everything, and the first route that matches a request
//  is the one that is run.
app.get("*", function(req, res) {
    res.send("YOU ARE A STAR!");
});

// REALLY IMPORTANT
// In Express, gotta tell the app to listen for different requests with app.listen()
// param1: Needs the port to listen on (PORT 3000 on an actual server, outside of C9)
// param2: Also needs the IP address of the server.
// param3: callback function to print (or something else), when the server starts.
app.listen(process.env.PORT, process.env.IP, function(){
  console.log("Server started!");
});

// NOTE: On C9, to visit the app in the browser:
//  0. run "node app.js" to start the server.
//  1. hit Preview -> Preview Running Application
//  2. copy the link of the new window that appears. 
```

## Routing Continued

* Show the `*` (pronounced "splat" or star") route matcher.
  * A catch-all or wildcard for remaining routes.

* Write routes containing route parameters.
> Consider Reddit. How likely is it that Reddit hard-codes its routes for every subreddit, every post, etc. ? Unlikely af! Instead, Reddit (and other sites) use route patterns.

  * How?
    * example:
      `app.get("/r/:subredditName" function(){});`
      * This syntax is saying match "the thing after "/r/" (but before another "/").
      * The thing is a variable.
      * Note the colon, that lets Express know it is a variable.
      
    * req.params holds the route parameters.
      
* What are route parameters/ route variables/path variabes?
  * Route variables are constructs in place to preserve a url pattern's structure.
    * for example, a user might naviate to a subreddit, but we can't make `if-else` statements for all posible choices of subreddit they could choose, that'd be ridiculous.
    * Instead, if we let out url pattern be "/r/:subredditName", we can just search the DB for the subredditName the user entered.
* Discuss route order.
  * order of routes matter (routes diplay short-circuit behavior)

## Express Boilerplate, v1

Steps
1. Create Project Directory and `cd` into it.
2. Run `npm init` wizard for a bunch of setup stuff (including dependencies setup).
    * Important: this line must go before installing something with the `--save` flag.
      * follow the directions pls
    * `init` should generate a package.json file.

3. Run `npm install express --save` to install express, and to add express to the dependencies.
    * **NOTE:** A package-losck.json might appear, it is safe to commit. It **SHOULD** be committed.
    * **NOTE:** a node_modules folder will probably be added, this is OK.
    
4. `touch` the app.js file (while inside your Project Directory).

5. Inside the app.js file, make sure it looks like this to start:
   ```javascript
   express = require("express");
   app = express();
   .
   .
   .
   . 
   .
   //OPTIONAL: You may want to add a splat route matcher for an error page
   // MUST be the last route in the file though. 
   app.get("*", function(req, res){
    res.send("ERROR 404: PAGE NOT FOUND!");
});

   // Make sure to set your app to listen, right at the very end of the app.js file.
   // tHiS iS wHat RuNs ThE sErVeR.
   // Also, NOTE: things starting w/ "process.env.*" are only used on Cloud9.
   //   WHen using an actual server,make sure to put its real port and IP address.
   app.listen(process.env.PORT, process.env.IP, function(){
    console.log("Now serving your app!");
});
   ``` 
   **NOTE:** the `.` (dots) represent where the majority of user written code (routes, etc) will go.
